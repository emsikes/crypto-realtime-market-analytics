# Catalog and schema setup
Create Unity Catallog structure for Crypto Realtime Analytics Pipeline

Layers:

* Bronze - Raw data from Kafka (via CoinGecko)
* Silver - Cleaned and deduplicated
* Gold - Aggregated metrics

```sql
SHOW EXTERNAL LOCATIONS;
```

```sql
CREATE CATALOG IF NOT EXISTS crypto_analytics
MANAGED LOCATION 'abfss://demo@deacourseextdlst.dfs.core.windows.net/crypto_analytics';
```

```sql
CREATE SCHEMA IF NOT EXISTS crypto_analytics.bronze;
CREATE SCHEMA IF NOT EXISTS crypto_analytics.silver;
CREATE SCHEMA IF NOT EXISTS crypto_analytics.gold;
```
